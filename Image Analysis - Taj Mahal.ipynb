```python
from skimage .transform import rotate
tajr = rotate (tajc, angle = 45)
imshow(tajr)
```

```python
from numpy import fliplr, flipud
tajflr = fliplr(tajc)
tajfud = flipud(tajc)
```

```python
from skimage import exposure
tajb = exposure.adjust_gamma(tajc, gamma = 0.25, gain =1)
imshow(tajb)
```

```python
tajd = exposure.adjust_gamma(tajc, gamma =2, gain =1)
imshow(tajd)
```

```python
tajds = exposure.adjust_gamma(tajc, gamma = 4, gain =1)
imshow(tajds)
```

```python
from skimage.filters import sobel_h, sobel_v
tajsobh = sobel_h(tajg)
imshow(tajsobh)
```

```python
tajsobv = sobel_v(tajsobh)
imshow(tajsobv)
```

```python
from skimage.filters import prewitt_h, prewitt_v
tajpbh =prewitt_h(tajg)
imshow(tajpbh)
```
